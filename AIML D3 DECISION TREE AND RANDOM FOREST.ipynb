#decesion tree alg to crime_rate 

```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
```

```python
df=pd.read_csv(r"C:\Users\ssaro\Downloads\demodt.txt")
```

```python
df
```

```python
df.isnull().sum()
```

```python
ind=df[['Literacy','Cleanliness','Crime_Rate']]
dep=df['Good']
```

```python
DT=DecisionTreeClassifier()
DT.fit(ind,dep)
```

```python
a=int(input("enter Literacy  :"))
b=int(input("enter Cleanlines"))
c=int(input("enter Crime_rate"))
out=DT.predict([[a,b,c]])
if out==1:
    print("good state")
else:
    print("bad state")
```

#decision tree alg to diabetes

```python
#diabetes(DT)
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
```

```python
df=pd.read_csv(r"C:\Users\ssaro\Downloads\diabetcsv2.csv")
df
```

```python
df.isnull().sum()
```

```python
ind=df[['preg','plas','pres','skin','insu','mass','pedi','age']]
dep=df['class']
```

```python
y=float(input("enter preg:"))
x=int(input("enter plasma:"))
z=float(input("enter blood pressure:"))
a=float(input("enter skin color:"))
b=float(input("enter insulin:"))
p=float(input("enter your mass:"))
q=float(input("enter peddy:"))
r=float(input("enter your age:"))
out=DT.predict([[y,x,z,a,b,p,q,r]])
print("the predicted class is",out[0])
```
